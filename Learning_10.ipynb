# Pivot Table

```python
import pandas as pd

Bangladesh1={
    'Date':['07/08/2022', '07/08/2022', '07/08/2022'],
    'City':['Dhaka','Chittagong','Khulna'],
    'Temperature':[34,36,42],
    'Humidity':[85,62,71]
}
Bangladesh_weather1=pd.DataFrame(Bangladesh1)
Bangladesh2={
    'Date':['21/08/2022', '21/08/2022','21/08/2022'],
    'City':['Dhaka','Chittagong','Khulna'],
    'Temperature':[44,56,62],
    'Humidity':[87,64,78]
}
Bangladesh_weather2=pd.DataFrame(Bangladesh2)
Bangladesh3={
    'Date':['10/09/2022', '10/09/2022', '10/09/2022'],
    'City':['Dhaka','Chittagong','Khulna'],
    'Temperature':[39,31,47],
    'Humidity':[87,67,74]
}
Bangladesh_weather3=pd.DataFrame(Bangladesh3)

df=pd.concat([Bangladesh_weather1,Bangladesh_weather2,Bangladesh_weather3])
df
```

```python
df.pivot(index='Date',columns='City')
```

# Pivot Table

```python
Bangladesh={
    'Date':['07/08/2022','07/08/2022','08/08/2022','08/08/2022','07/08/2022','07/08/2022','08/08/2022','08/08/2022','07/08/2022','07/08/2022','08/08/2022','08/08/2022'],
    'City':['Dhaka','Dhaka','Dhaka','Dhaka','Chittagong','Chittagong','Chittagong','Chittagong','Khulna','Khulna','Khulna','Khulna'],
    'Temperature':[34,36,42,45,44,56,62,57,39,31,47,42],
    'Humidity':[85,62,71,68,87,64,78,71,55,82,67,75]
}
Bangladesh_weather=pd.DataFrame(Bangladesh)
Bangladesh_weather
```

```python
Bangladesh_weather.pivot_table(index='City',columns='Date')
```

```python
Bangladesh_weather.pivot_table(index='City',columns='Date',aggfunc='sum')
```

```python
Bangladesh_weather.pivot_table(index='City',columns='Date',margins=True)
```
