```python
df.set_index('ID', inplace=True)
df
```

# Creating groups based on a column

```python
g=df.groupby('Gender')
g
```

```python
for Gender, Gender_df in g:
    print(Gender)
    print(Gender_df)
```

```python
g.get_group('male')
```
