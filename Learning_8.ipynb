```python
import pandas as pd
india={
    'City':['Mumbai','Delhi','Shimla'],
    'Temperature':[32,46,38],
    'Humidity':[80,60,70]
}
India_weather=pd.DataFrame(india)
India_weather
```

```python
Bangladesh={
    'City':['Dhaka','Chittagong','Khulna'],
    'Temperature':[34,36,42],
    'Humidity':[85,62,71]
}
Bangladesh_weather=pd.DataFrame(Bangladesh)
Bangladesh_weather
```

```python
USA={
    'City':['New York','Chicago','Oriando'],
    'Temperature':[21,18,32],
    'Humidity':[68,75,65]
}
USA_weather=pd.DataFrame(USA)
USA_weather
```

# Concat multiple dataframe

```python
pd.concat([India_weather,Bangladesh_weather,USA_weather])
```

```python
df=pd.concat([India_weather,Bangladesh_weather,USA_weather],keys=['India','Bangladesh','USA'])
df.to_csv('weather.csv')
df
```

```python
df=pd.concat([India_weather,Bangladesh_weather,USA_weather],ignore_index=True)
df.to_csv('weather.csv')
df
```

```python
temp_df=pd.DataFrame({
    'City':['Dhaka','Chittagong','Khulna'],
    'Temperature':[34,36,42]
})
hum_df=pd.DataFrame({
    'City':['Dhaka','Chittagong','Khulna'],
    'Humidity':[85,62,71]
})
pd.concat([temp_df,hum_df],axis=1)
```

```python
temp_df=pd.DataFrame({
    'City':['Dhaka','Chittagong','Khulna'],
    'Temperature':[34,36,42]
},index=[0,1,2])
hum_df=pd.DataFrame({
    'City':['Khulna','Dhaka'],
    'Humidity':[71,85]
},index=[2,0])
pd.concat([temp_df,hum_df],axis=1)
```
